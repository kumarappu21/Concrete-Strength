```python
#import pandas to read the csv file from PC
import pandas as pd
df = pd.read_csv('concrete.csv')
df.head()
```

```python
# rename the columns to make task easy and readable
df.rename(columns={'Cement (component 1)(kg in a m^3 mixture)':'cement'}, inplace = True)
df.rename(columns={'Blast Furnace Slag (component 2)(kg in a m^3 mixture)':'slag'}, inplace = True)
df.rename(columns={'Fly Ash (component 3)(kg in a m^3 mixture)':'ash'}, inplace = True)
df.rename(columns={'Superplasticizer (component 5)(kg in a m^3 mixture)':'superplasticizer'}, inplace = True)
df.rename(columns={'Fine Aggregate (component 7)(kg in a m^3 mixture)':'FA'}, inplace = True)
df.rename(columns={'Age (day)':'age'}, inplace = True)
df.head()
```

```python
x = df.drop(['slag', 'superplasticizer', 'age', 'fck', 'ash'], axis='columns')
x.head()
```

```python
y = df['fck']
y.head()
```

```python
from sklearn.preprocessing import PolynomialFeatures
degree = 2
poly = PolynomialFeatures(degree, include_bias=False)
x_poly = poly.fit_transform(x)
y_poly = y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test = train_test_split(x_poly,y_poly,test_size=0.2)
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,lr.predict(x_test))
print(r2)
```
